```python
import ikpy.chain
import numpy as np
import ikpy.utils.plot as plot_utils
```

```python
my_chain = ikpy.chain.Chain.from_urdf_file("./resources/arm.urdf")
```

```python
target = [0.05, 0.02, 0.03]
```

```python
print(my_chain.inverse_kinematics(target))
```

```python
real_frame = my_chain.forward_kinematics(my_chain.inverse_kinematics(target))
print("Computed position vector : %s, original position vector : %s" % (real_frame[:3, 3], target))
```

```python
%matplotlib widget
import matplotlib.pyplot as plt
fig, ax = plot_utils.init_3d_figure()
my_chain.plot(my_chain.inverse_kinematics(target), ax, target=target)
plt.xlim(-0.1, 0.1)
plt.ylim(-0.1, 0.1)
```
